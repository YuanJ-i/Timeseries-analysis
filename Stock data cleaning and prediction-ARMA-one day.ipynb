采用ARMA模型的基本步骤：

1.首先判断序列是否为平稳序列,这可以通过自相关图判断出来。如果自相关系数随着滞后(lags)迅速下降，就说明该序列平稳。即具备短期相关性，无长期相关性。

2.若不平稳，进行差分，一般一阶差分。

3.根据自相关图和偏自相关图来判断AR和MA阶数p,q. 一般p,q<5 

4.根据最小信息准则判断最优阶数： 比如AIC， AIC准则是赤池信息准则，该项准则运用下式的统计量评价模型的好坏：AIC=-2L/n+2K/n，其中L是对数似然值，n是观测值数目，k是被估计的参数个数，AIC的准则要求其越小越好。



```python
data['date_parsed'] = pd.to_datetime(data['Date'].astype(str)+' '+data['Time'],format="%Y%m%d %H:%M:%S")
data.index = data['date_parsed']
data.head(10)
```

```python
#prediction based on data on 20200701
stock_1 = data[data["Id"] == 1]["2020-07-01"]
stock_2 = data[data["Id"] == 2]["2020-07-01"]
stock_3 = data[data["Id"] == 3]["2020-07-01"]
stock_4 = data[data["Id"] == 4]["2020-07-01"]
#stock_1.shape
stock_1["NumTrades"].describe()
```

```python
# Now let's plot the total volume of stock being traded in 2020-07-01
plt.figure(figsize=(12, 12))
plt.subplots_adjust(top=1.25, bottom=1.2)
fig = plt.figure()
stocks = [stock_1, stock_2, stock_3, stock_4]
for i in range(1,5):
    plt.subplot(2, 2, i)
    stocks[i-1]["NumTrades"].plot(y = "NumTrades")
    plt.xlabel("Time")
    plt.ylabel("NumTrades")
    #plt.setp(fig.get_xticklabels(), rotation=30, horizontalalignment='right')
    fig.autofmt_xdate()
    plt.title("stock_"+str(i))
plt.tight_layout()
plt.show()
```

```python
#statsmodel里有绘制acf图的命令 很方便
#acf数值随着延迟下降，且趋近于0，这预示着日内时间序列具有短期相关性
#延迟t = 0, acf = 1; 

lags = [i for i in range(10)]

plot_acf(stock_1["NumTrades"], lags = lags, title = "Stock_1 Autocorrelation").show()
plot_pacf(stock_1["NumTrades"], lags = lags, title = "Stock_1 Partial Autocorrelation").show()

plot_acf(stock_2["NumTrades"], lags = lags, title = "Stock_2 Autocorrelation").show()
plot_pacf(stock_2["NumTrades"], lags = lags, title = "Stock_2 Partial Autocorrelation").show()

plot_acf(stock_3["NumTrades"], lags = lags, title = "Stock_3 Autocorrelation").show()
plot_pacf(stock_3["NumTrades"], lags = lags, title = "Stock_3 Partial Autocorrelation").show()

plot_acf(stock_4["NumTrades"], lags = lags, title = "Stock_4 Autocorrelation").show()
plot_pacf(stock_4["NumTrades"], lags = lags, title = "Stock_4 Partial Autocorrelation").show()
```

```python
#数据平滑
#数据差分--》成为稳定序列
stock_1["diff"] = stock_1["NumTrades"].diff(periods = 1)
stock_2["diff"] = stock_2["NumTrades"].diff(periods = 1)
stock_3["diff"] = stock_3["NumTrades"].diff(periods = 1)
stock_4["diff"] = stock_4["NumTrades"].diff(periods = 1)
```

```python
#inplace 原地操作 修改data
stock_1.dropna(how = 'any',inplace = True)
stock_2.dropna(how = 'any',inplace = True)
stock_3.dropna(how = 'any',inplace = True)
stock_4.dropna(how = 'any',inplace = True)
```

```python
#使用自回归模型预测
```

```python
'''
Use diff data: diff 
split sample into train and test data
'''
def AR_train(stock, string, p = 3):
    n_sample = stock.shape[0]
    #print(n_sample)
    n_train=int(0.9*n_sample)+1
    n_forecast=n_sample-n_train
    ts_train = stock['diff'][:n_train]
    ts_test = stock['diff'][n_train:]
    #print("training size: ",ts_train.shape)
    #print("testing size: ",ts_test.shape)
    #print("Training Series:", "\n", ts_train.tail(), "\n")
    #print("Testing Series:", "\n", ts_test.head())
    ar = AutoReg(ts_train, lags = p)
    ar_results = ar.fit()#fit模型
    #ar_results.summary()
    
    pred = ar_results.predict('20200701 09:59:00', '20200701 13:30:00',dynamic=True)#预测，指定起始与终止时间。预测值起始时间必须在原始数据中，终止时间不需要
    #print ("Prediction from Autoregression: ")
    fig1 = plt.figure()
    plt.title(string+" AR("+str(p)+") predicted first-order change per minute")
    stock['diff'].plot()
    pred.plot()
    fig1.show()
    
    #return pred
    #AR还原一阶差分
    start = stock['NumTrades'][n_train-1]
    #print(start)
    restore = pred.cumsum() + start 
    fig2 = plt.figure()
    plt.title(string+" AR("+str(p)+") prediction restored per minute")
    
    stock['NumTrades'].plot()
    restore.plot()
    fig2.show()
```

#AR model:

原理：实际上就是对历史数据的线性回归，历史的记忆深度取决于阶数p的取值，一般小于5.

可以看到AR的拟合基本上复现了最后几个数据点的波动和总体趋势。

由于AR，自回归模型就是简单的lag阶线性回归； 稳定时序只具有短期相关性， 故预测delta_t<lag的数据比较可行；大于lag， 预测能力显著下降

```python
#stock-1, autoregression
AR_train(stock_1, "stock-1", p = 5)
```

```python
#stock-2, autoregression
AR_train(stock_2, "stock-2", p = 5)
```

```python
#stock-3, autoregression
AR_train(stock_3, "stock-3", p = 5)
```

```python
#stock-4, autoregression
AR_train(stock_4, "stock-4", p = 5)
```

#ARMA model:

原理：相较于AR模型的拟合，ARMA(p,q)的预测更加平稳，波动更小

但总体上只能给出线性的预测,与AR的结果类似

阶数尝试了p,q<5的各种组合，总体上类似

```python
def ARIMA_train(stock, string, order = (1,0,1)):
    n_sample = stock.shape[0]
    #print(n_sample)
    n_train=int(0.9*n_sample)+1
    n_forecast=n_sample-n_train
    ts_train = stock['diff'][:n_train]
    ts_test = stock['diff'][n_train:]
    #print("training size: ",ts_train.shape)
    #print("testing size: ",ts_test.shape)
    #print("Training Series:", "\n", ts_train.tail(), "\n")
    #print("Testing Series:", "\n", ts_test.head())
    arima = ARIMA(ts_train, order=order)
    arima_results = arima.fit()#fit模型
    #ar_results.summary()
    
    pred = arima_results.predict('20200701 09:59:00', '20200701 13:30:00',dynamic=True)#预测，指定起始与终止时间。预测值起始时间必须在原始数据中，终止时间不需要
    #print ("Prediction from Autoregression: ")
    fig1 = plt.figure()
    plt.title(string+" AR("+str(order[0])+")MA("+str(order[2])+") predicted first-order change per minute")
    stock['diff'].plot()
    pred.plot()
    fig1.show()
    
    #return pred
    #AR还原一阶差分
    start = stock['NumTrades'][n_train-1]
    #print(start)
    restore = pred.cumsum() + start 
    fig2 = plt.figure()
    plt.title(string+" AR("+str(order[0])+")MA("+str(order[2])+") prediction restored per minute")
    
    stock['NumTrades'].plot()
    restore.plot()
    fig2.show()
```

```python
#stock-1, arima
ARIMA_train(stock_1, "stock-1", order = (3,0,2))
```

```python
#stock-2, arima
ARIMA_train(stock_2, "stock-2", order = (3,0,2))
```

```python
#stock-3, arima
ARIMA_train(stock_3, "stock-3", order = (3,0,2))
```

```python
#stock-4, arima
ARIMA_train(stock_4, "stock-4", order = (3,0,2))
```
